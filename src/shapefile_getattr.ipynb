```python
imp_fields = shp_gdf[['NAME_1', 'ID_1', 'NAME_2', 'ID_2']]
```

```python
imp_fields.head()
```

```python
def add_hyphon(string: str):
    if ' ' in string:
        return string.strip().replace(' ', '-')
    else:
        return string
    
```

```python
imp_fields[['NAME_1', 'NAME_2']] = imp_fields[['NAME_1', 'NAME_2']].applymap(add_hyphon).copy()
```
